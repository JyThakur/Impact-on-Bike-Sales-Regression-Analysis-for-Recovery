```python
X_train = day_train
y_train = day_train.pop('cnt')
```

## 5. Building a linear model

### (A) <font color='orange'>RFE(Recursive Feature Elimination)

```python
# importing packages
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with top 20 features
lm = LinearRegression()
lm.fit(X_train,y_train)

rfe = RFE(lm, 20)
rfe = rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

### (B) <font color='orange'>Building model using statsmodel, for the detailed statistics</font>
#### Usign REF and p-value
##### Model 1

```python
# Creating X_test dataframe with RFE selected variables
X_train_1M = X_train[col]
```

```python
# import packages
import statsmodels.api as sm 
```

```python
# Adding a constant variable 
X_train_1M = sm.add_constant(X_train_1M)
```

```python
lm_1 = sm.OLS(y_train,X_train_1M).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm_1.summary())
```

#### <font color=red> Observations </font>
- Feature `holiday` `summer` `February` and `May` have insignificantly high p-value i.e. >0.05, so we can drop them and rebulid new model

```python
# dropping highly correlated variables
X_train_1D = X_train_1M.drop(['holiday','summer','February','May'],axis=1)
```

##### Model 2

```python
# Adding a constant variable
X_train_2M = sm.add_constant(X_train_1D)

lm_2 = sm.OLS(y_train,X_train_2M).fit()   # Running the linear model

# Printing summary of model
print(lm_2.summary())
```

### <font color='orange'>Calculating VIFs, for Model 1</font>

```python
# dropping const
X_train_Vif1 = X_train_2M.drop(['const'], axis=1)
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['Features'] = X_train_Vif1.columns
vif['VIF'] = [variance_inflation_factor(X_train_Vif1.values, i) for i in range(X_train_Vif1.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

#### <font color=red> Observations </font>
- Features `hum` `temp` and `workingday` significantly have high VIF value(i.e. >5) and including them in model could lead to muliticollinearity; so it is important to drop them and rebuild new model

```python
# dropping highly correlated variables
X_train_2D = X_train_Vif1.drop(['hum','temp','workingday'],axis=1)
```

##### Model 3

```python
# Adding a constant variable
X_train_3M = sm.add_constant(X_train_2D)
```

```python
lm_3 = sm.OLS(y_train,X_train_3M).fit()   # Running the linear model
```

```python
# Printing summary of model
print(lm_3.summary())
```

#### <font color=red> Observations </font>
- Features `July` `winter` `Saturday` seems to be highly insignificant as their p-value is >0.05, so we can drop them and rebuild new model 

```python
# dropping highly insignificant variables
X_train_3D = X_train_3M.drop(['July','Saturday','winter'],axis=1)
```

##### Model 4

```python
# Adding a constant variable
X_train_4M = sm.add_constant(X_train_3D)

lm_4 = sm.OLS(y_train,X_train_4M).fit()   # Running the linear model

# Printing summary of model
print(lm_4.summary())
```

### <font color='orange'>Calculating VIFs, for Model 4</font>

```python
# dropping const from Model 3
X_train_Vif2 = X_train_4M.drop(['const'], axis=1)
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['Features'] = X_train_Vif2.columns
vif['VIF'] = [variance_inflation_factor(X_train_Vif2.values, i) for i in range(X_train_Vif2.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

#### <font color=red> Observations </font>
- As it is visible from stats that VIF and p-value of all the 10 Features are under accaptable range i.e. VIF<=5 and p-value<0.05.

### (D) <font color='orange'> Residual Analysis

```python
y_train_cnt = lm_4.predict(X_train_4M)
```

```python
# Importing the required libraries for plots.
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_cnt), bins = 20)
fig.suptitle('Residual of Final Model', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
plt.show()
```

#### <font color=red> Observations </font>
- The error makes Normal Distribution i.e. mean of error/Residual is 0.0, so we can proceed for Prediction using Test set.
